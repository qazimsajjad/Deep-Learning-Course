```python
def show(trainImages):
    plt.figure(figsize=(10,10))
    _, axis = plt.subplots(4,4)
    for i, ax in enumerate(axis.ravel()):
        ax.imshow(trainImages[i], 'gray')
        ax.axis('off')
    plt.show()
        
def visualize(x, y, functionName):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=functionName, color='blue')
    plt.title(f'{functionName.title()} Activation Function')
    plt.xlabel('x')
    plt.ylabel(f'{functionName.title()}(x)')
    plt.axhline(0.5, color='red', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
```
