```python
import cv2
import keras
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from keras.models import Sequential
from keras.models import load_model
from keras.losses import SparseCategoricalCrossentropy
import matplotlib.pyplot as plt
import numpy as np
from keras.optimizers import Adam
from keras.datasets import mnist
from keras.layers import Softmax
from keras.layers import ReLU, LeakyReLU
from keras.activations import sigmoid, tanh, elu, selu, softplus, swish

from keras.callbacks import TensorBoard
```

```python
def show(trainImages):
    plt.figure(figsize=(10,10))
    _, axis = plt.subplots(4,4)
    for i, ax in enumerate(axis.ravel()):
        ax.imshow(trainImages[i], 'gray')
        ax.axis('off')
    plt.show()
        
def visualize(x, y, functionName):
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=functionName, color='blue')
    plt.title(f'{functionName.title()} Activation Function')
    plt.xlabel('x')
    plt.ylabel(f'{functionName.title()}(x)')
    plt.axhline(0.5, color='red', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()
```

```python
(trainImages, trainLabels), (testImages, testLabels) = mnist.load_data()
```

```python
show(trainImages)
```

```python
trainImages.shape, testImages.shape
```

```python
model = Sequential()

model.add(Conv2D(6, kernel_size=(3,3), input_shape=(28,28,1), activation='relu', padding='same'))
model.add(MaxPooling2D((2,2))) # None, 13x13x6
model.add(Conv2D(12, kernel_size=(3,3), activation='relu', padding='same')) # None, 11x11x12
model.add(Conv2D(24, kernel_size=(3,3), activation='relu'))  # None, 3x3x24
model.add(Conv2D(36, kernel_size=(3,3), activation='relu')) # None,
model.add(Conv2D(48, kernel_size=(3,3), activation='relu')) # None, 18x18x48
model.add(MaxPooling2D(2,2))


model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

## TensorBoard callback

```python
tb_callback = TensorBoard(log_dir='logs/', histogram_freq=1)
```

```python
model.compile(loss='SparseCategoricalCrossentropy', optimizer = 'adam', metrics=['accuracy',])
```

```python
epochs = 10
batchSize = 128
```

```python
records = model.fit(trainImages, trainLabels, epochs=epochs, batch_size=batchSize, callbacks = [tb_callback])
```

```python
model.evaluate(testImages, testLabels)
```

```python
preds = model.predict(testImages)
len(preds)
```

## TensorBoard

TensorBoard is a powerful visualization tool that comes with TensorFlow, designed to help you understand, debug, and optimize your machine learning models. 
### TensorBoard Different Functionalities:
`Scalars:` Visualize scalar metrics such as loss and accuracy over time during training.

`Graphs:` Visualize the computational graph of your model.

`Histograms:` Visualize the distributions of weights and biases in your model's layers.

`Distributions:` Visualize the distribution of activations across different layers of your model.

`Embeddings:` Visualize high-dimensional embeddings in lower-dimensional space for tasks like visualizing word embeddings.

`Profile:` Profile your TensorFlow program's execution to identify performance bottlenecks.
